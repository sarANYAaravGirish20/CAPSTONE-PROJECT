#                                   Human Rights violation

```python
import pandas as pd
import numpy as np
import geopandas as gpd

import matplotlib.pyplot as plt
%matplotlib inline

import warnings
warnings.filterwarnings('ignore')
```

```python
# Reading the csv file
hr = pd.read_csv('35_Human_rights_violation_by_police_2001_2010.csv')
hr.head()
```

```python
g1 = pd.DataFrame(hr.groupby(['Area_Name'])['Cases_Registered_under_Human_Rights_Violations'].sum().reset_index())
g1.columns = ['State/UT','Cases Reported']
```

```python
g1.head()
```

```python
plt.figure(figsize=(10, 5))
plt.xticks(rotation=40,size=8)
plt.bar(g1['State/UT'].values, g1['Cases Reported'].values)
```

# Cases registered against Police under Human Rights Violation

```python
g3 = pd.DataFrame(hr.groupby(['Year'])['Cases_Registered_under_Human_Rights_Violations'].sum().reset_index())
g3.columns = ['Year','Cases Registered']
```

```python
plt.figure(figsize=(8, 5))
plt.bar(g3['Year'].values, g3['Cases Registered'].values)
plt.xlabel('Year')
plt.ylabel('Cases Registered')
plt.title('Cases under Human Rights Violations')
plt.show()
     
```

```python
g4 = pd.DataFrame(hr.groupby(['Year'])['Policemen_Chargesheeted','Policemen_Convicted'].sum().reset_index())
```

```python
g4
```

```python
plt.figure(figsize=(10, 5))
width=0.3
plt.bar(g4['Year'].values,g4['Policemen_Chargesheeted'].values,width=width)
plt.bar(g4['Year'].values + width, g4['Policemen_Convicted'].values,width=width)
plt.legend(['Policemen Chargesheeted', 'Policemen Convicted'])
plt.show()
```

#                            PROPERTY STOLEN AND RECOVERED CASE

```python
# Reading the CSV file
t = pd.read_csv('11_Property_stolen_and_recovered_nature_of_property_2001_2010.csv')
t.head()
```

```python
g= pd.DataFrame(t.groupby(['Area_Name'])['Cases_Property_Stolen','Group_Name'].sum().reset_index())
```

```python
g.head()
```

```python
g = pd.DataFrame(t.groupby(['Group_Name'])['Value_of_Property_Stolen'].sum().reset_index())
g
```

```python
t1 =t['Cases_Property_Stolen'].sum()
t2 = t['Cases_Property_Recovered'].sum()
```

```python
cases = ['Cases_Property_Stolen','Cases_Property_Recovered']
values = [t1,t2]
```

```python
plt.figure(figsize=(8, 5))
plt.pie(values,labels=cases, autopct='%1.1f%%')
plt.axis('equal')
plt.legend(cases)
plt.title('vehicles stolen and recovered')
plt.tight_layout()
plt.show()
```

Almost 74 percent of cases for property theft but the property recovered is 26 %
