#                        CRIME AGAINST SC YEAR WISE

```python
plt.figure(figsize=(20,12))
plt.subplot(2,4,1)
df1.groupby("Year")["Murder"].sum().plot(title="Murder")
plt.subplot(2,4,2)
df1.groupby("Year")["Rape"].sum().plot(title="Rape")
plt.subplot(2,4,3)
df1.groupby("Year")["Kidnapping and Abduction"].sum().plot(title="Kidnapping and Abduction")
plt.subplot(2,4,4)
df1.groupby("Year")["Dacoity"].sum().plot(title="Dacoity")
plt.subplot(2,4,5)
df1.groupby("Year")["Robbery"].sum().plot(title="Robbery")
plt.subplot(2,4,6)
df1.groupby("Year")["Arson"].sum().plot(title="Arson")
plt.subplot(2,4,7)
df1.groupby("Year")["Hurt"].sum().plot(title="Hurt")
plt.subplot(2,4,8)
plt.show()
```

Above graph reveals that over the years the number of reported cases of Hurt , Atrocity , Rape and other Crimes are increased whereas the rate of reporting other crimes has stayed the almost same


Despite governments best effort the number of atrocities and hurt cases are increasing over the years. Rajasthan ,Uttarpradesh , Bihar ,Maharashtra and Rajasthan seem to be hotspot for crimes against Scs.
