#  KNN

```python
from sklearn.model_selection import train_test_split
train_x, test_x, train_y, test_y = train_test_split(train_x, df_train, test_size = 0.2, random_state=42)
from sklearn.neighbors import KNeighborsRegressor
neigh = KNeighborsRegressor(n_neighbors=3)
neigh.fit(train_x,train_y)
KNeighborsRegressor(n_neighbors=3)
pred_y=neigh.predict(test_x)
pred_y
```

```python
from sklearn.metrics import r2_score
knn=r2_score(test_y,pred_y)
knn
```

```python
neigh1 = KNeighborsRegressor(n_neighbors=5)
neigh1.fit(train_x,train_y)
```

```python
pred_y1=neigh1.predict(test_x)
knn1=r2_score(test_y,pred_y1)
knn1
```

# RANDOM FOREST REGRESSOR

```python
from sklearn.ensemble import RandomForestRegressor
regressor1 = RandomForestRegressor(n_estimators = 100, random_state = 42)
regressor1.fit(train_x,train_y)
RandomForestRegressor(random_state=42)
y_pred_rfr = regressor1.predict(test_x)
rfr=r2_score(test_y,y_pred_rfr)
rfr
```
